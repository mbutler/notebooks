```python
from google_images_download import google_images_download
from fastai.vision import verify_images,Path
```

```python
category = 'undead'
path = Path(category)
```

```python
max_images = 5
```

```python
# classes = ['toilet', 'skeleton']
```

optional cell to convert a list of string items into a python list of classes

```python
text_list = """aberation
celestial
construct
dragon
elemental
fey
fiend
giant
humanoid
monstrosity
ooze
plant
swarm
undead
"""
classes = text_list.splitlines( )
```

```python
keywords = ",".join(classes)
```

```python
modifiers = "d&d monster"
```

If multi_images is true, it will dump all downloads into the same folder with the category name.
Usefull if there are multiple google searches per class

```python
multi_images = True
```

```python
response = google_images_download.googleimagesdownload()
```

```python
arguments = {"keywords":keywords,"suffix_keywords":modifiers,"limit":max_images,"print_urls":False,"output_directory":category,"no_directory":multi_images}
```

```python
response.download(arguments)
```

clean up any images fastai can't open correctly

```python
for c in classes:
    print(c)
    verify_images(path/c, delete=True, max_size=500)
```
